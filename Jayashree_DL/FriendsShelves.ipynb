```python
###  method 2 by pasting url of book cover from web

import pytesseract
pytesseract.pytesseract.tesseract_cmd = r"C:\\Program Files\\Tesseract-OCR\\tesseract.exe"
from PIL import Image
import pandas as pd
import requests
from io import BytesIO
# Create a DataFrame
df=pd.read_csv('book_data.csv')
df1 = df.iloc[:10]
df2 = pd.DataFrame(df1)


def extract_text_from_image(image_url):
    try:
        response = requests.get(image_url)
        response.raise_for_status()
        image = Image.open(BytesIO(response.content))
        extracted_text = pytesseract.image_to_string(image)
        return extracted_text.strip().lower().split('\n')
    except Exception as e:
        print("Error:", str(e))
    return []

def search_dataframe_by_text(dataframe, search_text, column):
    matching_rows = dataframe[dataframe[column].str.contains(search_text, case=False)]
    return matching_rows

def main():
    image_url = input("Enter the image URL: ")
    extracted_texts = extract_text_from_image(image_url)

    if extracted_texts:
        print("Texts extracted from the image:", extracted_texts)

        # Collect matching rows for each extracted text
        matching_rows_all = []
        for text in extracted_texts:
            # Perform a search in the DataFrame based on the extracted text
            matching_rows = search_dataframe_by_text(df2, text, 'text')
            if not matching_rows.empty:
                matching_rows_all.append((text, matching_rows))

        if matching_rows_all:
            print("Matching rows for each extracted text:")
            for text, matching_rows in matching_rows_all:
                print(f"Text: {text}")
                print("Matching rows:")
                print(matching_rows)
        else:
            print("No matching rows found for any extracted text.")
    else:
        print("No text extracted from the image.")

if __name__ == "__main__":
    main()
```
